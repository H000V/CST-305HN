```python
import time
from scipy.integrate import solve_ivp

def ode_system(x, y):
    return [-((3 / x) + y[0]) / (3 * x * y[0])]

def run_rkf_simulation(x_span, y0, rtol=1e-3, atol=1e-6):
    start_time = time.perf_counter()
    sol = solve_ivp(ode_system, x_span, y0, method='RK45', rtol=rtol, atol=atol)
    end_time = time.perf_counter()
    
    print(f"Settings\t: Range {x_span}, Initial y = {y0[0]}")
    print(f"Tolerances\t: rtol = {rtol}, atol = {atol}")
    print(f"Steps taken\t: {len(sol.t) - 1}")
    print(f"Evaluations\t: {sol.nfev}")
    print(f"Time\t\t: {end_time - start_time:.6f} seconds")
    print(f"Final y\t\t: {sol.y[0][-1]:.6f}\n")

print("Standalone ODE Solve")
run_rkf_simulation(x_span=(1.0, 1.5), y0=[1.0])
print("Altered Initial Condition")
run_rkf_simulation(x_span=(1.0, 1.5), y0=[5.0])
print("Tighter Tolerances")
run_rkf_simulation(x_span=(1.0, 1.5), y0=[1.0], rtol=1e-8, atol=1e-10)
```
